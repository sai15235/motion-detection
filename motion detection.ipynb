```python
vid=cv2.VideoCapture(0)

suc1,fr1 =vid.read()
frg1=cv2.cvtColor(fr1,cv2.COLOR_BGR2GRAY)

while True:
    suc,fr=vid.read()
    frg=cv2.cvtColor(fr,cv2.COLOR_BGR2GRAY)

    diff=cv2.subtract(frg1,frg)/255

    if np.sum(diff)>1000:
        cv2.putText(frg1,"motion detected",(10,100),cv2.FONT_ITALIC,3,(255))
    
    cv2.imshow("live camera",frg1)    

    if cv2.waitKey(1) == ord("q"):
      break

    frg1=frg
    
cv2.destroyAllWindows()
```

# detection in roi

```python
import cv2
import numpy as np


vid = cv2.VideoCapture(0)
suc1, fr1 = vid.read()
frg1 = cv2.cvtColor(fr1, cv2.COLOR_BGR2GRAY)

roi = (100,100,200,200)
x1, y1, x2, y2 = roi
while True:
    # Read the current frame
    suc, fr = vid.read()
    frg = cv2.cvtColor(fr, cv2.COLOR_BGR2GRAY)
  
    diff = cv2.subtract(frg1, frg)

    # Threshold the diff image to get motion areas
    _,thresh = cv2.threshold(diff, 25, 255, cv2.THRESH_BINARY)

    # Calculate motion in ROI
    roi_thresh = thresh[y1:y2, x1:x2]
    motion_score = np.sum(roi_thresh) / 255  # Count of changed pixels

    # Draw the ROI rectangle on the live color frame
    cv2.rectangle(fr, (x1, y1), (x2, y2), (0, 0, 255), 5)


    if motion_score> 1000:  
        cv2.putText(fr, "Motion Detected in ROI", (10, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2)

    cv2.imshow("Live Camera with Motion Detection", fr)

  
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

   
    frg1 = frg

cv2.destroyAllWindows()
```
